```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import os
```

```python
traffic_signs_path = './traffic_signs'
bag_img_path = './img_GiratoireDroite'

traffic_signs_files = os.listdir(traffic_signs_path)
bag_img_files = os.listdir(bag_img_path)
```

```python
traffic_signs = []
bag_imgs = []
traffic_signs_label = []

for traffic_sign_filename in traffic_signs_files:
        img = cv.imread(traffic_signs_path+'/'+traffic_sign_filename)
        img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
        img = cv.resize(img, (int(img.shape[1]/2), int(img.shape[0]/2)), interpolation=cv.INTER_AREA)
        traffic_signs.append(img)
        traffic_signs_label.append(traffic_sign_filename.split('.')[0])

for bag_img_filename in bag_img_files:
        img = cv.imread(bag_img_path+'/'+bag_img_filename)
        img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
        img = cv.resize(img, (int(img.shape[1]/2), int(img.shape[0]/2)), interpolation=cv.INTER_AREA)
        bag_imgs.append(img)
```

```python
fig, axes = plt.subplots(8,8,figsize=(20,20))
for i in range(64):
    axes[i%8][i//8].imshow(traffic_signs[i])
    axes[i%8][i//8].axis('off')
plt.show()
```

```python
fig, axe = plt.subplots(1,1, figsize=(15,15))
axe.imshow(bag_imgs[0])
axe.axis('off')
plt.show()
```
